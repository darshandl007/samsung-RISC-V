## 4.1 Why Functions?

Imagine you need to calculate energy consumed in 10 different places in your code. Without functions, you copy-paste the formula 10 times. If the formula changes, you update it 10 times and likely miss one.

With a function, you write the formula once, give it a name, and call it wherever needed. When the formula changes, you fix it in one place.

Functions let you:
- **Reuse** code without repeating it
- **Name** a concept so code reads like plain English
- **Isolate** one job into one place — easier to test and fix

## 4.2 Defining a Function

```python
def function_name(parameter1, parameter2):
    # code here
    return result
```

- `def` — keyword that starts a function definition
- `function_name` — the name you give it (same rules as variables)
- `parameters` — inputs the function receives (optional)
- `return` — sends a value back to whoever called the function (optional)

````python
def calculate_energy(power_kw, hours):
    energy = power_kw * hours
    return energy
````

Defining a function does not run it. You must **call** it:

````python
result = calculate_energy(1.5, 2)
print(result)   # 3.0

result2 = calculate_energy(1.5, 4)
print(result2)  # 6.0
````

## 4.3 Parameters vs Arguments

- **Parameter** — the name in the function definition: `power_kw`, `hours`
- **Argument** — the actual value you pass when calling: `1.5`, `2`

They are often called the same thing informally. Just know: parameters are placeholders, arguments are the real values.

## 4.4 Functions Without a Return Value

A function does not have to return anything. It can just perform an action:

````python
def print_sensor_reading(temperature, humidity, is_occupied):
    print(f"Temperature : {temperature}°C")
    print(f"Humidity    : {humidity}%")
    print(f"Occupied    : {is_occupied}")
    print("-" * 30)

print_sensor_reading(32, 65.4, True)
print_sensor_reading(27, 70.0, False)
````

## 4.5 Default Parameter Values

You can give parameters a default value. If the caller does not provide that argument, the default is used:

````python
def calculate_energy(power_kw, hours=1):
    return power_kw * hours

print(calculate_energy(1.5))     # uses default hours=1 → 1.5
print(calculate_energy(1.5, 3))  # overrides default → 4.5
````

## 4.6 Functions Calling Other Functions

Functions can call other functions. This is how the pipeline in the project works:

````python
def calculate_energy(power_kw, hours):
    return power_kw * hours

def should_ac_be_on(temperature, is_occupied, threshold=30):
    return is_occupied and temperature > threshold

def run_one_cycle(temperature, humidity, is_occupied):
    ac_on = should_ac_be_on(temperature, is_occupied)

    if ac_on:
        energy = calculate_energy(power_kw=1.5, hours=1/60)
    else:
        energy = 0.0

    return ac_on, energy


# Run one cycle
ac_status, energy_used = run_one_cycle(temperature=33, humidity=65, is_occupied=True)
print(f"AC: {ac_status}  |  Energy: {energy_used:.4f} kWh")
````

## 4.7 Variable Scope

Variables created inside a function **only exist inside that function**. They are called **local variables**.

````python
def calculate_energy(power_kw, hours):
    energy = power_kw * hours   # 'energy' is local to this function
    return energy

result = calculate_energy(1.5, 2)
print(result)   # 3.0

# print(energy)  # This would cause a NameError — 'energy' does not exist here
````
