## 3.1 Making Decisions with if/else

Programs need to make decisions based on conditions. In Python, you use `if` and `else`.

**Syntax:**
```python
if condition:
    # code runs when condition is True
else:
    # code runs when condition is False
```

**Important:** The indented block (4 spaces) is what belongs to the `if`. Python uses indentation to define blocks of code — do not skip it.

````python
temperature = 32

if temperature > 30:
    print("AC ON")
else:
    print("AC OFF")
````

Change `temperature` to `25` and run again. The output changes because the condition `25 > 30` is `False`.

## 3.2 Comparison Operators

These operators compare two values and return `True` or `False`:

| Operator | Meaning | Example |
|---|---|---|
| `==` | equal to | `temperature == 30` |
| `!=` | not equal to | `status != "OFF"` |
| `>` | greater than | `temperature > 30` |
| `<` | less than | `humidity < 80` |
| `>=` | greater than or equal | `temperature >= 30` |
| `<=` | less than or equal | `occupancy <= 0` |

**Note:** `=` assigns a value. `==` compares two values. These are different.

````python
temperature = 30

print(temperature > 30)   # False — 30 is not greater than 30
print(temperature >= 30)  # True  — 30 is greater than OR equal to 30
print(temperature == 30)  # True  — 30 equals 30
````

## 3.3 elif — Multiple Conditions

When you have more than two possible outcomes, use `elif` (short for "else if"):

````python
temperature = 26

if temperature > 30:
    print("AC ON — cooling required")
elif temperature > 25:
    print("Fan ON — mild cooling")
else:
    print("All devices OFF — comfortable")
````

Python checks conditions from top to bottom and stops at the first `True` one.

## 3.4 Logical Operators — Combining Conditions

Use `and`, `or`, `not` to combine multiple conditions:

| Operator | Meaning |
|---|---|
| `and` | Both conditions must be True |
| `or` | At least one condition must be True |
| `not` | Reverses True to False and vice versa |

````python
temperature = 32
is_occupied = True

# AC should only run if room is occupied AND temperature is high
if is_occupied and temperature > 30:
    print("AC ON")
else:
    print("AC OFF")
````

````python
temperature = 32
is_occupied = False

# Room is empty — AC stays off even if hot
if is_occupied and temperature > 30:
    print("AC ON")
else:
    print("AC OFF — room is empty")
````

## 3.5 The while Loop

A `while` loop repeats a block of code **as long as** a condition is `True`.

```python
while condition:
    # code repeats here
```

This is how the simulation in `main.py` keeps running cycle after cycle.

````python
cycle = 1

while cycle <= 5:
    print(f"Simulation cycle {cycle}")
    cycle = cycle + 1   # increase cycle by 1 each time

print("Simulation complete")
````

**Warning:** If the condition never becomes `False`, the loop runs forever. This is called an **infinite loop**. Always make sure something inside the loop eventually makes the condition `False`.

**Shorthand:** `cycle = cycle + 1` can also be written as `cycle += 1`.

````python
cycle = 1

while cycle <= 5:
    print(f"Cycle {cycle}")
    cycle += 1  # same as cycle = cycle + 1
````

## 3.6 The for Loop

A `for` loop repeats over a sequence — a list of items, or a range of numbers.

````python
# range(1, 6) generates numbers 1, 2, 3, 4, 5
for cycle in range(1, 6):
    print(f"Cycle {cycle}")
````

````python
# Looping over a list of values
sensor_readings = [28, 31, 33, 27, 30]

for reading in sensor_readings:
    if reading > 30:
        print(f"{reading}°C — AC ON")
    else:
        print(f"{reading}°C — AC OFF")
````

## 3.7 break — Stopping a Loop Early

````python
cycle = 1
max_energy = 10.0
total_energy = 0.0

while True:  # runs forever UNLESS we break
    total_energy += 1.5
    print(f"Cycle {cycle} — Total energy: {total_energy} kWh")
    cycle += 1

    if total_energy >= max_energy:
        print("Energy limit reached. Stopping.")
        break  # exits the loop immediately
````

## 3.8 Putting It Together — A Mini Simulation

This combines everything from Chapters 2 and 3:

````python
import random  # built-in Python module for generating random numbers

TEMP_THRESHOLD = 30
total_energy = 0.0
ac_power_kw = 1.5
cycle_duration_hours = 1/60  # one minute per cycle

for cycle in range(1, 6):
    # Simulate a random temperature between 24 and 36 degrees
    temperature = random.randint(24, 36)
    is_occupied = random.choice([True, False])

    # Decision logic
    if is_occupied and temperature > TEMP_THRESHOLD:
        ac_on = True
        energy = ac_power_kw * cycle_duration_hours
        total_energy += energy
    else:
        ac_on = False
        energy = 0.0

    print(f"Cycle {cycle}: Temp={temperature}°C  Occupied={is_occupied}  AC={ac_on}  Energy={energy:.4f} kWh")

print(f"\nTotal energy consumed: {total_energy:.4f} kWh")
````

Run this cell multiple times. Each run produces different output because `random.randint()` generates different numbers each time. This is exactly how the project's `SensorSimulator` works.
