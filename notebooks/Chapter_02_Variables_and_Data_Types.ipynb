## 2.1 What is a Variable?

A variable is a named container that holds a value. You give it a name, assign a value, and use that name throughout your code.

Think of it like a labeled box:  
- The label is the variable name  
- The contents are the value

You create a variable using `=` (assignment operator):

````python
temperature = 32
print(temperature)
````

You can change the value stored in a variable at any time:

````python
temperature = 32
print(temperature)

temperature = 28
print(temperature)
````

## 2.2 Naming Rules

Variable names in Python must follow these rules:

| Rule | Example |
|---|---|
| Start with a letter or underscore | `temperature`, `_value` |
| Can contain letters, numbers, underscores | `room_temp_2` |
| Cannot start with a number | `2temp` — **invalid** |
| Cannot contain spaces | `room temp` — **invalid** |
| Case-sensitive | `Temperature` and `temperature` are different variables |

**Convention:** Use `snake_case` — all lowercase, words separated by underscores. Example: `room_temperature`, `is_occupied`.

## 2.3 The Four Data Types Used in This Project

Every value in Python has a **type**. The type determines what you can do with the value.

This project uses four types:

### int — Integer (whole numbers)

````python
temperature = 32          # degrees Celsius
occupancy_count = 4       # number of people in the room
cycle_number = 1          # simulation cycle

print(int(temperature))  # <class 'int'>
````

### float — Floating Point (decimal numbers)

````python
humidity = 65.4           # percentage
energy_consumed = 1.5     # kilowatt-hours
ac_power_rating = 1.2     # kilowatts

print(float(humidity))     # <class 'float'>
````

### str — String (text)

````python
device_status = "ON"      # or "OFF"
room_name = "Living Room"
alert_message = "High temperature detected!"

print(str(device_status)) # <class 'str'>
````

### bool — Boolean (True or False)

Booleans represent yes/no, on/off, true/false states. Note: `True` and `False` are capitalized in Python.

````python
is_occupied = True        # is the room occupied?
ac_on = False             # is the AC currently on?
alert_active = False      # is there an active alert?

print(bool(is_occupied))  # <class 'bool'>
````

## 2.4 Using Variables Together

````python
power_kw = 1.5      # AC power in kilowatts
hours = 2.0         # hours of operation

energy = power_kw * hours
print(energy)       # 3.0 kWh
````

## 2.5 f-Strings — Printing Variables with Text

An f-string lets you embed variable values directly inside a string. Put `f` before the opening quote and wrap variable names in `{}`.

````python
temperature = 32
humidity = 65.4
is_occupied = True

print(f"Temperature: {temperature}°C")
print(f"Humidity: {humidity}%")
print(f"Room occupied: {is_occupied}")
````

You can also do arithmetic inside the `{}`:

````python
power_kw = 1.5
hours = 3

print(f"Energy consumed: {power_kw * hours} kWh")
````

## 2.6 Type Conversion

Sometimes you need to convert between types:

````python
temperature = 32
temperature_as_float = float(temperature)   # 32 → 32.0
temperature_as_string = str(temperature)    # 32 → "32"

print(temperature_as_float)
print(temperature_as_string)
print(str(temperature_as_string))
````
