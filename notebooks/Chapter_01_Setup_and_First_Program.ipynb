## 1.1 What is Python?

Python is a programming language. A programming language lets you give instructions to a computer in a way both you and the computer can understand.

You write instructions in a `.py` file (or in a notebook like this one). Python reads those instructions and executes them.

Python is used in:
- Web development
- Data science and AI
- IoT and embedded systems
- Automation

This tutorial uses Python to simulate a smart energy system — sensors, decision logic, and a live dashboard.

## 1.2 What is a Terminal?

A terminal (also called command line or shell) is a text-based interface to your computer. Instead of clicking, you type commands.

**On Mac:** Press `Cmd + Space`, type `Terminal`, press Enter.  
**On Windows:** Press `Win + R`, type `cmd`, press Enter.

Common commands you will use:

| Command | What it does |
|---|---|
| `python --version` | Check which Python version is installed |
| `pip install <name>` | Install a Python package |
| `cd foldername` | Move into a folder |
| `ls` (Mac) / `dir` (Windows) | List files in current folder |
| `python filename.py` | Run a Python file |

## 1.3 Installing Python

1. Go to https://python.org
2. Click **Download Python 3.x** (use version 3.10 or higher)
3. Run the installer
   - **Important on Windows:** Check the box that says **"Add Python to PATH"** before clicking Install

After installation, open your terminal and type:

```bash
python --version
```

You should see something like: `Python 3.11.4`  
If you see an error, Python did not install correctly — repeat the steps above.

## 1.6 Python as a Calculator

Python can do arithmetic. Try running these:

````python
print(10 + 5)
print(30 - 8)
print(4 * 7)
print(100 / 4)
````

These are the four basic operators: `+` `-` `*` `/`

Two more useful ones:

````python
print(10 ** 2)   # 10 to the power of 2 = 100
print(10 % 3)    # remainder when 10 is divided by 3 = 1
````

The `#` symbol starts a **comment**. Python ignores everything after `#` on the same line. Comments are notes for humans reading the code.

## 1.7 What Happens When Code Has an Error?

Errors are normal. Every programmer sees them daily. Run the cell below:

````python
print("This line works")
print("This line has no quotes")   # SyntaxError — Python does not understand this
````

Python will show a **SyntaxError**. Read error messages carefully — they tell you what went wrong and on which line.

Fix the error in the cell above and run it again.
